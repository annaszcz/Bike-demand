```python
# Checking bike demand by week day
import calendar
import pandas as pd


#Defining a function that uses calendar.weekday to get the day of the week
def findDayFromTimestamp(timestamp):
    year = timestamp.year
    month = timestamp.month
    day = timestamp.day
    day_number = calendar.weekday(year, month, day)
    
    # Days list starting from Monday (since calendar.weekday() returns 0 for Monday)
    days = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]
    
    return days[day_number]

# function findDayFromTimestamp to 'Date' column and create a new column 'day_of_week'
daily_w['day_of_week'] = daily_w['Date'].apply(findDayFromTimestamp)
hr_w['day_of_week'] = hr_w['Date'].apply(findDayFromTimestamp)
```
